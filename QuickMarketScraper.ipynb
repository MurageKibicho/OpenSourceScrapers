```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd

webPages = [
    "https://www.quickmart.co.ke/collections/beverages-1",
    "https://www.quickmart.co.ke/collections/body-care-make-up",
    "https://www.quickmart.co.ke/collections/confectioneries",
    "https://www.quickmart.co.ke/collections/quick-dairy",
    "https://www.quickmart.co.ke/collections/electronics",
    "https://www.quickmart.co.ke/collections/flour-salt-sugar",
    "https://www.quickmart.co.ke/collections/health-wellness",
    "https://www.quickmart.co.ke/collections/juices-carbonated-drinks",
    "https://www.quickmart.co.ke/collections/meat-products-eggs",
    "https://www.quickmart.co.ke/collections/rice-cereals",
    "https://www.quickmart.co.ke/collections/soaps-detergents",
    "https://www.quickmart.co.ke/collections/spreads-food-additives"    
]

#Open Page
driver = "/home/murage/Desktop/bb/Prices/geckodriver"
browser = webdriver.Firefox(executable_path=driver)
actions = ActionChains(browser)
website = webPages[1]
browser.implicitly_wait(20) 
browser.get(website)
browser.implicitly_wait(20) 
names = []
prices = []

##Quickmart devs are crazy
##print(names[0].text +" "+ prices[0].text)
##print(names[1].text +" "+ prices[2].text)
##print(names[2].text +" "+ prices[4].text)
##print(names[3].text +" "+ prices[6].text)
##print(names[4].text +" "+ prices[8].text)
#for i in(range(len(names))):
#    print(names[i].text + " "+ prices[i*2].text)
browser.implicitly_wait(20)
while True:
    try:
        temp1 = browser.find_elements_by_class_name('productitem--title')
        temp2 = browser.find_elements_by_class_name('productitem--price ')
        for i in(range(len(temp1))):
            names.append(temp1[i].text)
            prices.append(temp2[i*2].text)
        WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "pagination--next"))).click()
    except TimeoutException:
        break
        
for i in(range(len(names))):
    print(names[i] + " "+ prices[i])
    


```
