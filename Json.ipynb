```python
import requests
import json
import subprocess

#Make initial get reqest
response = requests.get('https://apis.odibets.com/v4/matches?src=&sport_id=&tab=&country_id=&day=&sort_by=&competition_id=&trials=0')

error = 0.1
programName = "./m.o"

#Check if get request successful
if response.status_code == 200 and 'application/json' in response.headers.get('Content-Type',''):
    #load json into nested dictionary
    dictionaryResult = json.loads(response.content.decode('utf-8'))
    #Get individual competition IDs
    for eachCompetition in dictionaryResult["data"]["competitions"]:
        competitionID = eachCompetition["competition_id"]
        #Get url for each competition ID
        newUrl = "https://apis.odibets.com/v4/matches?src=&sport_id=soccer&tab=snapshot&country_id=&day=0&sort_by=&competition_id=" +competitionID+"&trials=0"
        newResponse = requests.get(newUrl)
        if newResponse.status_code == 200 and 'application/json' in newResponse.headers.get('Content-Type',''):
            newDictionary = json.loads(newResponse.content.decode('utf-8'))
            for eachNewCompetition in newDictionary["data"]["competitions"]:
                try:
                    for eachMatch in eachNewCompetition["matches"]:
                        print(eachMatch["home_team"] + " " + eachMatch["away_team"])
                        for eachMarket in eachMatch["markets"]:
                            commandList = []
                            commandList.append(programName)
                            for eachOutcome in eachMarket["outcomes"]:
                                commandList.append(eachOutcome["odd_value"])
                                #print(eachOutcome["odd_value"] + " ",end = "")
                            commandList.append("%s"%(error))
                            print(commandList)
                            out = subprocess.Popen(commandList,stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
                            stdout,stderr = out.communicate()
                            print(stdout) 
                            print("\n")
                            break #avoid 1x2
                except KeyError:
                    print("Failed here")
                        
                    
                    
        
```
